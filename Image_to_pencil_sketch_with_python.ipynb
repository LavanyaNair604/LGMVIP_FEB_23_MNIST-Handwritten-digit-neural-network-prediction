# **Importing libraries**

**We use google patches here as google colab dosent support cv2 imshow**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

## **Reading the image**

```python
image = cv2.imread("/content/Valorant-Jett.png")
```

## **This is the image I downloaded from internet**

```python
cv2_imshow(image)
```

## **Converting it to RGB**

```python
RBG_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RBG_image)
```

## **Convert it into grayscale**

```python
greyscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(greyscale)
```

## **Inverting the grayscale image**

```python
inverting = 255-greyscale
```

## **Lets see the image now**

```python
plt.imshow(inverting)
```

```python
blurred = cv2.GaussianBlur(inverting, (21, 21), 0)
```

## **The blurred image**

```python
plt.imshow(blurred)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(greyscale, inverted_blurred, scale=256.0)
```

## **The inverted blurred image**

```python
plt.imshow(inverted_blurred)
```

## **Original one with matplotlib**

```python
plt.imshow(image)
```

## **Pencil sketch one with matplotlib**

```python

plt.imshow(pencil_sketch)
```

## **Original and pencil sketch with cv2**

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
